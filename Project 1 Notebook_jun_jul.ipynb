```python
import pandas as pd
```

```python
jun_data = pd.read_csv("Project1_Resources/FlightData_Jun2019.csv")
jun_filter_dest = jun_data.loc[jun_data['Dest']=='MCI']
jun_filter_origin = jun_data.loc[jun_data['Origin']=='MCI']
# jul_data = pd.read_csv("Project1_Resources/FlightData_Jul2019.csv")
jun_filter_dest
```

```python
jun_mci_dest = jun_filter_dest[['Month','DayofMonth','DayOfWeek',
                                 'Marketing_Airline_Network',
                                 'Origin',
                                 'OriginCityName',
                                 'Dest',
                                 'DestCityName',
                                 'DepDelayMinutes',
                                 'ArrDelayMinutes',
                                 'Cancelled',
                                 'CarrierDelay',
                                 'WeatherDelay',
                                 'NASDelay',
                                 'SecurityDelay',
                                 'LateAircraftDelay']]

jun_mci_origin = jun_filter_origin[['Month','DayofMonth','DayOfWeek',
                                 'Marketing_Airline_Network',
                                 'Origin',
                                 'OriginCityName',
                                 'Dest',
                                 'DestCityName',
                                 'DepDelayMinutes',
                                 'ArrDelayMinutes',
                                 'Cancelled',
                                 'CarrierDelay',
                                 'WeatherDelay',
                                 'NASDelay',
                                 'SecurityDelay',
                                 'LateAircraftDelay']]

jun_mci_origin
jun_mci_dest

merged_jun_data = pd.concat([jun_mci_origin,jun_mci_dest], axis=0)
merged_jun_data
```

```python
merged_jun_data.to_csv("Jun_mci_data.csv",index=False)
```
